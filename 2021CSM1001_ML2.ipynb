```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.colors
import seaborn as sns
import math 
```

```python
def w_calc(X, y, n,m):
  phi = []
  for i in range(n):
    row = []
    for j in range(m):
      row.append(x[i]**(j))
    phi.append(row)
  phi = np.array(phi)
  phi_T = phi.transpose()
      
  res = np.matmul(phi_T,phi)  
      
  res2 = np.linalg.inv(res)
      
  res3 = np.matmul(res2,phi_T)
      
  w = np.matmul(res3,y)
      
  return w
```

```python
def predic(x, n_points, degree, weight):
  final_y = []
  for i in range(n_points):
    val =0
    for j in range(degree):
      val= val + (weight[j]*(x[i]**(j)))
    final_y.append(val)
  #print(final_y)
  return final_y
```

```python
d=[1,2,3,5,7,10]

n_points = 20
x = np.linspace(0,2*np.pi,n_points)
y=(np.cos(2*np.pi*x)+ x/(2*np.pi)+np.random.normal(0,0.004))


erms_l = []

erms_d = []

fig, axs = plt.subplots(2,3, figsize=(20, 10))
fig.subplots_adjust(hspace = .5, wspace=.5)
axs = axs.ravel()
count=0
      
for degree in d:
  weights = w_calc(x, y, n_points, degree)
  weight = weights.tolist()
  axs[count].scatter(x, y, c="blue")
  y_pred = predic(x, n_points, degree, weight)
  axs[count].plot(x,y_pred,c="red")
  axs[count].set_title('M = %i' %degree)
  ew = (np.sum((y_pred - y)**2))/n_points
  erms = np.sqrt((2*ew)/n_points)
  erms_l.append(erms)
  count = count + 1
```

**Observations:**


*   We can see that **when M=1**, the model gives us a **poor representation** of our y function and also **poorly predicts** the y values.
*   As we keep on **increasing M**, the model starts becoming **similar** to our **y function** and also **starts fitting** the data points.


*  **M=10** gives us the **fittest curve** here and the curve also starts looking **similar to our y function**.






```python
sns.barplot(x=d, y=erms_l, palette="husl",lw=3,ec="black",hatch="..")
plt.xlabel('M')
plt.ylabel('ERMS')
plt.title('Data points = 20')
erms_d.append(np.mean(erms_l))
```

**Observations:**


*   The above barplot graph represents Erms values to different M values when our data points are 20.
*   We can see as we keep **on increasing M** our **Erms also keeps decreasing** because our curve starts fitting the data points much nicely.
